## 2.2: Variables

A **variable** is a labeled box that stores a value so you can reuse it later. You create one with `=` (read it as "gets", not "equals").

````python
name = "Ada"
year_founded = 2015
years_running = 2026 - year_founded

print(name)
print(years_running)
````

A few rules for variable names:
- Use lowercase with underscores: `total_enrollments`, not `TotalEnrollments` or `x2`
- Names can't start with a number or contain spaces
- **Make names descriptive.** `completion_rate` beats `cr` every time. You will thank yourself in three weeks when you reread your own code

Variables can be **updated**. The box keeps whatever you put in it last:

````python
score = 10
print(score)

score = score + 5   # take the current value, add 5, store it back
print(score)
````

## 2.3: Data types

Every value in Python has a **type**. The four you need today:

| Type | Name | Examples |
|---|---|---|
| `int` | Integer (whole number) | `5`, `-12`, `2026` |
| `float` | Decimal number | `3.14`, `0.5`, `-2.0` |
| `str` | String (text) | `"hello"`, `"42"` |
| `bool` | Boolean (True/False) | `True`, `False` |

You can check any value's type with the `type()` function:

````python
print(type(5))
print(type(5.0))
print(type("5"))
print(type(True))
````

**Why types matter:** Python treats types differently, and mixing them up is the #1 source of beginner errors. Watch what happens with numbers vs. strings:

````python
print(5 + 5)       # int + int → math
print("5" + "5")   # str + str → gluing text together ("concatenation")
````

Same symbols, completely different behavior. This next one **fails on purpose**. Run it and read the error message at the bottom; the last line is the important part:

````python
print("5" + 5)   # str + int → Python refuses
````

**Errors are normal.** You'll see hundreds of them. The real skill is reading the last line of the message calmly. Here Python tells you it can't combine (`concatenate`) a `str` and an `int`.

You can convert between types when you need to:

````python
enrollment_text = "250"                 # a string, maybe from a spreadsheet export
enrollment_number = int(enrollment_text)  # convert it to an integer

print(enrollment_number + 50)             # now math works
print(str(300) + " participants")         # or convert a number to a string
````

## 2.4: Basic math

Python is a very fancy calculator. The operators:

| Operator | Meaning | Example | Result |
|---|---|---|---|
| `+` | Add | `7 + 3` | `10` |
| `-` | Subtract | `7 - 3` | `4` |
| `*` | Multiply | `7 * 3` | `21` |
| `/` | Divide | `7 / 3` | `2.333...` |
| `**` | Power | `7 ** 2` | `49` |
| `//` | Divide, drop the remainder | `7 // 3` | `2` |
| `%` | Remainder ("modulo") | `7 % 3` | `1` |

Normal order of operations applies, and parentheses work like in math class.

````python
completed = 187
enrolled = 250

completion_rate = completed / enrolled
print(completion_rate)

# Multiply by 100 and round to 1 decimal place for a percentage
print(round(completion_rate * 100, 1))
````

One more trick before the finale: **f-strings**. Put an `f` before the quotes and you can drop variables directly into text with `{curly braces}`. This is the cleanest way to build readable output, and we'll use it constantly:

````python
completed = 187
enrolled = 250
rate = round(completed / enrolled * 100, 1)

print(f"{completed} of {enrolled} participants completed the course ({rate}%)")
````
